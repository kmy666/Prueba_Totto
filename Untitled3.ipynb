```python
import pandas as pd

# Cargar datos
df = pd.read_excel('C:/Users/Milo/Downloads/PRUEBA 2 P.SAC (2).xlsx') 

# Filtrar solo maletas de viaje (si hay otros productos)
# Eliminar filas con precios/costos negativos o cero
df = df[(df['VTAFINAL CON DESCUENTO'] > 0) & (df['UNDS COSTEADAS'] > 0)]
```

```python
df.describe()
```

```python
# Convertir fechas y definir períodos
df['FECHA'] = pd.to_datetime(df['FECHA'], dayfirst=True)

# Período PROMOCIÓN (9 jun - 9 jul)
promo_inicio = pd.to_datetime('09-06-2023', dayfirst=True)
promo_fin = pd.to_datetime('09-07-2023', dayfirst=True)

# Período POST-PROMOCIÓN (10 jul - 10 ago)
post_inicio = pd.to_datetime('10-07-2023', dayfirst=True)
post_fin = pd.to_datetime('10-08-2023', dayfirst=True)


df['Periodo'] = 'Otro'
df.loc[df['FECHA'].between(promo_inicio, promo_fin), 'Periodo'] = 'Promocion'
df.loc[df['FECHA'].between(post_inicio, post_fin), 'Periodo'] = 'Post_Promocion'


df = df[df['Periodo'].isin(['Promocion', 'Post_Promocion'])]
# Agrupar por factura (IMPORTANTE para margen exacto)
facturas = df.groupby(['NumFactura', 'Periodo']).agg({
    'VTAFINAL CON DESCUENTO': 'sum',
    'UNDS COSTEADAS': 'sum',
    'CantUnidadesVendidas': 'sum',
    'COM-SET VIAJE JUNIO': lambda x: 'APLICA' if 'APLICA' in x.values else 'NO APLICA'
}).reset_index()


facturas['Margen_Absoluto'] = facturas['VTAFINAL CON DESCUENTO'] - facturas['UNDS COSTEADAS']
facturas['Margen_%'] = (facturas['Margen_Absoluto'] / facturas['UNDS COSTEADAS']) * 100
```

```python

resumen = facturas.groupby('Periodo').agg({
    'VTAFINAL CON DESCUENTO': 'sum',
    'CantUnidadesVendidas': 'sum',
    'Margen_Absoluto': 'sum',
    'Margen_%': 'mean'
})

print(resumen)
```

```python
from scipy import stats


margen_promo = facturas[facturas['Periodo']=='Promocion']['Margen_%']
margen_post = facturas[facturas['Periodo']=='Post_Promocion']['Margen_%']


t_stat, p_val = stats.ttest_ind(margen_promo, margen_post)
print(f"\nDiferencia en márgenes: p-value = {p_val:.4f}")
print("¿Significativo?" , "Sí" if p_val < 0.05 else "No")
```
